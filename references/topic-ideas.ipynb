## Data source

Der Datensatz zeigt Transaktionsdaten eines E-Commerce Unternehmen aus dem Vereinigten Koenigreich. Die Daten wurden zwischen dem 01.12.2010 und dem 09.12.2011 erfasst.

Mithilfe des UCI Machine Learning Repository wurde der Datensatz von wurde Dr Daqing Chen zur Verfuegung gestellt. <br>

## Data characterisitcs

Jeder Beobachtungswert stellt innerhalb des Datensatzes eine Transaktion / Produktbestellung dar, wobei mehrere Transaktionen / Produktbestellungen einer Rechnung(snummer) zugeteilt werden kann.

Der Datensatz enthaelt 541909 Beobachtungswerte mit 8 Spalten.

| Spaltenname  | Beschreibung  | Skalenniveau |
|---|---|---|
|InvoiceNo | Rechnungsnummer, 6-stellige Nummer, welche jede Transaktion eindeutig zuordnet. <br>Falls die Nummer ein "c" enthält wurde die Transaktion storniert. | Nominal |
|StockCode | Produktnummer, 5-stellige Nummer, welche jedem einzelnem Produkt eindeutig zugeordnet ist. | Nominal |
|Description | Beschreibung des Produkts. | Nominal |
|Quantity | Menge der einzelnen Produkte pro Transaktion. | Verhaeltnis |
|InvoiceDate | Uhrzeit und Datum der Rechnung. | Intervall |
|UnitPrice | Preis pro Produkt. | Verhaeltnis |
|CustomerID | Kundennummer, 5-stellige Nummer, welche je einem Kunden eindeutig zugeordnet ist. | Nominal |
|Country | Name des Landes in welchem der Kunde ansässig ist. | Nominal |
