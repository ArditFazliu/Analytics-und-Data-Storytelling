```python
#Setup

import pandas as pd
import altair as alt
from pathlib import Path
```

```python
parent_path = str(Path().resolve().parent.parent) + "/"
data_path = "data/"
raw_path = "raw/"
raw_data_file = "raw_data.csv"
```

```python
df = pd.read_csv(parent_path + data_path + raw_path + raw_data_file, encoding= 'unicode_escape')
```

```python
# Erstellen einer neuen Variable, um den gesamten Preis pro Transaktion zu ermitteln
df["TotalPrice"] = df["Quantity"] * df["UnitPrice"]
```

```python
#Fehlende Werte auslesen
df.isna().sum()
```

```python
#Fehlende Werte löschen
df.dropna()
```

```python
non_stockcode = []
for s in df["StockCode"]:
    if any(i.isdigit() for i in s) == False:
        non_stockcode.append(s)
```

```python
pd.value_counts(non_stockcode)
```

```python
set_stockcode = set(non_stockcode)
df = df[~df["StockCode"].isin(set_stockcode)]
```

```python
processed_path = "processed/"
processed_file = "processed_data.csv"
df.to_csv(parent_path + data_path + processed_path + processed_file, index=False)
```
