```python
#Setup
import pandas as pd
import altair as alt
from pathlib import Path
import numpy as np
import statsmodels.api as sm
alt.data_transformers.disable_max_rows()
```

```python
parent_path = str(Path().resolve().parent.parent) + "/"
data_path = "data/"
processed_path = "processed/"
processed_data_file = "processed_data.csv"
```

```python
df = pd.read_csv(parent_path + data_path + processed_path + processed_data_file)
```

```python
df
```

```python
df["CustomerID"] = df["CustomerID"].astype(str).str[:-2]
df = df.dropna(subset=["CustomerID"])
df = df[df["CustomerID"] != "n"]
df = df.sort_values(by='CustomerID')
```

```python
chart = alt.Chart(df.head(5000)).mark_circle().encode(
    x='CustomerID',
    y='TotalPrice',
    tooltip=['CustomerID', 'TotalPrice']
).interactive()

chart = chart.transform_filter(
    (alt.datum.TotalPrice < 800) & (alt.datum.TotalPrice > -200)
)

chart
```

```python
new_df = df.groupby('CustomerID')['TotalPrice'].sum().reset_index()

# Umbenennen der Spalten
new_df.rename(columns={'TotalPrice': 'Sum_TotalPrice'}, inplace=True)
```

```python
#Delete all customers with negative total price and total price > 40000
new_df = new_df[new_df["Sum_TotalPrice"] > 0]
new_df = new_df[new_df["Sum_TotalPrice"] < 40000]
```
